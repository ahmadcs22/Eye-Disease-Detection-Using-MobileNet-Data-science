```python
import os
from PIL import Image

dataset_path = r'C:\Users\Ahmad\Downloads\archive\dataset'  # your dataset path

# Minimum size allowed
min_width = 50
min_height = 50

# Target size for MobileNetV2 
target_size = (224, 224)

for folder_name in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder_name)
    
    if os.path.isdir(folder_path):
        for filename in os.listdir(folder_path):
            file_path = os.path.join(folder_path, filename)
            
            try:
                img = Image.open(file_path)
                img = img.convert('RGB')  # Convert to RGB (in case it was grayscale)

                # Check if the image is too small
                if img.size[0] < min_width or img.size[1] < min_height:
                    print(f"Image too small, deleting: {file_path}")
                    os.remove(file_path)
                    continue
                
                # Resize the image
                img = img.resize(target_size)
                
                # Save back the resized image
                img.save(file_path)
            
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
                os.remove(file_path)  # If error, delete the image
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2, Xception
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam

# 1. DATA SPLIT (80% train / 20% val)
dataset_path = r'C:\Users\Ahmad\Downloads\archive\dataset'  # your root folder with 4 subfolders

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_gen = datagen.flow_from_directory(
    dataset_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical',
    subset='training',
    shuffle=True
)
val_gen = datagen.flow_from_directory(
    dataset_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# helper to build a transfer-learning model
def build_model(base_model_class, base_name):
    base = base_model_class(
        weights='imagenet',
        include_top=False,
        input_shape=(224,224,3)
    )
    base.trainable = False
    model = models.Sequential([
        base,
        layers.GlobalAveragePooling2D(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(train_gen.num_classes, activation='softmax')
    ], name=base_name)
    model.compile(
        optimizer=Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

Model Training

```python
# 2. MOBILENETV2
mobilenet_model = build_model(MobileNetV2, 'MobileNetV2_ft')
print(mobilenet_model.summary())

mobilenet_history = mobilenet_model.fit(
    train_gen,
    epochs=10,
    validation_data=val_gen
)
```

Model Evaluation

```python
# evaluate
mnv2_val_loss, mnv2_val_acc = mobilenet_model.evaluate(val_gen)
print(f"MobileNetV2 — val_loss: {mnv2_val_loss:.4f}, val_acc: {mnv2_val_acc:.4f}")
```

Model Deployment

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load the saved model
model = load_model('eye_disease_mobilenet_model.h5')

# Class names (make sure in same order as during training)
class_names = ['cataract', 'diabetic_retinopathy', 'glaucoma', 'normal']
```

```python
def predict_image(img_path):
    # Load and preprocess the image
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = img_array / 255.0  # Rescale like during training
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

    # Predict
    prediction = model.predict(img_array)
    predicted_class = class_names[np.argmax(prediction)]

    # Plot image and prediction
    plt.imshow(img)
    plt.title(f"Predicted: {predicted_class}")
    plt.axis('off')
    plt.show()

    return predicted_class
```
